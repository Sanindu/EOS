```python
example_file = glob.glob(r"/content/drive/MyDrive/Dataset/DATA_HARE242100N3303300UTM15R_2_vhr.tif")[0]
plt.show(example_file)
cm_gray = plt.get_cmap()
im = imread(example_file, as_gray=True)
plt.imshow(im, cmap=cm_gray)
print("InputImage : RBC.jpg")

plt.show()
```

```python


blobs_log = blob_log(im, max_sigma=30, num_sigma=10, threshold=.4)

blobs_log[:, 2] = blobs_log[:, 2] * sqrt(2)
numrows = len(blobs_log)
print("Number of Buildings: " ,numrows)

fig, ax = plt.subplots(1, 1)
plt.imshow(im, cmap=cm_gray)
for blob in blobs_log:
    y, x, r = blob
    c = plt.Circle((x, y), r+5, color='lime', linewidth=2, fill=False)
    ax.add_patch(c)
plt.show()
```
